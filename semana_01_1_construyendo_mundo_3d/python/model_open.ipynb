Import the libraries

```python
import trimesh
import numpy as np
import matplotlib as plot
import os
```

Define the path of the model

```python
path = "pikachu\scene.gltf"
```

Load the 3D model

```python
model = trimesh.load(path)
```

Show the loaded model

```python
model.show()
```

This cell count the number of vertices, edges and faces for all the elements in the scene

```python
for name,malla in model.geometry.items(): #We go through all the item in the scene
    print(name)
    print("Vrtices:", len(malla.vertices))
    print("Aristas:", len(malla.edges_unique))
    print("Caras:", len(malla.faces))
```

In the next cell is highlighted the edges of the first (and biggest) mesh in color red 

```python
mesh = list(model.geometry.values())[0] #Select the first object (mesh)
edges = trimesh.load_path(mesh.vertices[mesh.edges_unique]) #Select only the edges of the mesh
#Assing the red color to every edge in the mesh
edges.colors = np.tile(
    [255, 0, 0, 255],
    (len(edges.entities), 1)
)

edges.show() #Show the edges
```

We can see that Pikachu has no face, so probably the first mesh correspond to the body and the others are the face and the rest of the details 

Now we can load every mesh, asing a different color for every one, and show all in the same plot

```python
#Second Mesh
mesh2 = list(model.geometry.values())[1] 
edges2 = trimesh.load_path(mesh2.vertices[mesh2.edges_unique]) 
edges2.colors = np.tile(
    [0, 255, 0, 255],
    (len(edges2.entities), 1)
)

#Third Mesh
mesh3 = list(model.geometry.values())[2] 
edges3 = trimesh.load_path(mesh3.vertices[mesh3.edges_unique]) 
edges3.colors = np.tile(
    [0, 0, 255, 255],
    (len(edges3.entities), 1)
)

#Fourth Mesh
mesh4 = list(model.geometry.values())[3] 
edges4 = trimesh.load_path(mesh4.vertices[mesh4.edges_unique]) 
edges4.colors = np.tile(
    [0, 255, 255, 255],
    (len(edges4.entities), 1)
)

all_edges = trimesh.Scene([edges, edges2, edges3, edges4])
all_edges.show()
```
